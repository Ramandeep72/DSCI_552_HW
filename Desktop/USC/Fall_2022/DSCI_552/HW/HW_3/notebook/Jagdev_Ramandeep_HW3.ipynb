# Ciii) standard deviation of each time domain features with bootstrapped method to build 90% bootstrapped confidence interval for each feature

```python
higher_confidence_interval = []
lower_confidence_interval  = []

for i in range(len(my_list)):
    samples = df[my_list[i]]
    a = np.asarray(samples)
    samples = a
    bootResult = bs.bootstrap(samples, stat_func=bs_stats.std, alpha=0.1)
    
    higher_confidence_interval.append(bootResult.upper_bound)
    
    lower_confidence_interval.append(bootResult.lower_bound)
    
```

```python
std_dev_1 = df.std()
std_dev_1 = list(std_dev_1)
```

```python
std_dev_with_bounds = pd.DataFrame({'Feature Name': my_list,"Lower Bound": lower_confidence_interval, 'Standard Deviation':std_dev_1, 'Upper Bound': higher_confidence_interval})
```
